```python
import pandas as pd
```

```python
swetest = "./swe_unix_src_2.10.03/src/swetest"
exportPath = './export/'
marketPath = './market/'
imgPath = './image/'

planet = (
    # 'helio'
    'geo'
    # '1' # moon
    # '3' # venus
    # '4' # mars
)
astro_df = pd.read_csv(marketPath + planet + 'Astro.csv',parse_dates = True,index_col=0)

columns = astro_df.columns
row = lambda x: print(x)
list(map(row, columns))
```

```python
futr_exog_list = astro_df \
.filter(regex='^((?!\(|Selena).)*$') \
.filter(regex='Earth|Sun|Mercury|Venus|Moon|Mars|Jupiter|Saturn|Uranus|Neptune|Pluto|Chiron|Pholus|Ceres|Pallas|Juno|Vesta|sirius|true') \
.filter(regex='declination|longitude|distance|speed|sirius')

# "declination":        # major pivots, but not good enough
# "longitude":          # most pivots, but can't tell top or bottom
# "latitude":           # similar to declination
# "distance":           # sometimes it's spot on, sometimes it's off
# "speed":              # sometimes it's spot on, sometimes it's off    
# "right_asc":          # negative tendencies
# "meridian_dst":       # by itself, no
# "zenith_dst":         # fair
# "azimuth":            # negative tendencies
# "altitude":           # no
# "ecliptical":         # similar to declination
# "equatorial":         # no
# "uv_ecliptical":      # no
# "uv_equatorial":      # no
# "house_pos":          # similar to longitude
# "house_num":          # by itself, no
# "zodiac":             # fair
# "sirius_dec":         # bullish bias
# "sirius_long":        # too choppy, but idea is there
    
# futr_exog_list = unwrap_df(futr_exog_list)        
    
print(futr_exog_list.columns)

file = 'slim_' + planet + 'Astro.csv'
out = exportPath + file
futr_exog_list.to_csv(out,index=True)  
```

```python
df = futr_exog_list

from matplotlib import pyplot as plt
first_column_name = df.columns[0]
plt.plot(df.iloc[:, 0].head(365*10))
plt.xlabel(first_column_name)
plt.show()
```
